# ACC Basketball Homework Assigment

### Reading the CSV File

```python
import pandas as pd

# Load the CSV file into a DataFrame
acc_basketball_df = pd.read_csv('/content/acc_players-2324F.csv')

# Display the first 5 rows of the dataset
print("ACC Basketball Stats:")
acc_basketball_df.head()
```

```python
# Check the structure of the DataFrame
print("Data Info")
acc_basketball_df.info()
```

```python
# Check for missing values
print("Missing Values in Data Info:")
acc_basketball_df.isnull().sum()
```

```python
# Get statistical summaries of the data
print("Statistical Summary :")
acc_basketball_df.describe()
```

```python
# Rename columns
acc_basketball_df.rename(columns={
    'Unnamed: 0': 'Rk',
    'Unnamed: 1': 'Player',
    'Unnamed: 2': 'Class',
    'Unnamed: 3': 'Pos',
    'Unnamed: 4': 'School',
    'Totals': 'G',
    'Unnamed: 6': 'MP',
    'Unnamed: 7': 'TRB',
    'Unnamed: 8': 'AST',
    'Unnamed: 9': 'STL',
    'Unnamed: 10': 'BLK',
    'Unnamed: 11': 'TOV',
    'Unnamed: 12': 'PF',
    'Unnamed: 13': 'PTS',
    'Shooting': 'FG%',
    'Unnamed: 15': '2P%',
    'Unnamed: 16': '3P%',
    'Unnamed: 17': 'FT%',
    'Advanced': 'PER',
    'Unnamed: 19': 'WS',
    'Unnamed: 20': 'BPM'
}, inplace=True)

print("Updated ACC Basketball Info Columns:")
print(acc_basketball_df.columns)
```

```python
acc_basketball_df
```

```python
# Check for missing values
print("Missing Values in Data Info:")
acc_basketball_df.isnull().sum()
```

```python
# Select only the columns that should be numeric
numeric_cols = ['Rk','G', 'MP', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'FG%', '2P%', '3P%', 'FT%', 'PER', 'WS', 'BPM']

# Convert the selected columns to integers and fill NaN values with 0
for col in numeric_cols:
  acc_basketball_df[col] = pd.to_numeric(acc_basketball_df[col], errors='coerce')

for col in numeric_cols:
    acc_basketball_df[col] = acc_basketball_df[col].fillna(0)

print(acc_basketball_df.info())
```

```python
# Check for missing values
print("\nMissing Values in Data Info:")
acc_basketball_df.isnull().sum()
```

### Basic Analysis

```python
# Calculate the total points scored by all players combined
total_points = acc_basketball_df['PTS'].sum()
print(f"Total points scored by all players: {total_points}")
```

```python
# Find the player who has played the most minutes (MP)
most_minutes_player = acc_basketball_df.loc[acc_basketball_df['MP'].idxmax()]
print(f"Player with the most minutes: {most_minutes_player['Player']}, Minutes: {most_minutes_player['MP']}")
```

```python
# Identify the top 5 players in terms of total rebounds (TRB)
top_5_rebounds = acc_basketball_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print("Top 5 players by total rebounds:")
print(top_5_rebounds)
```

### Player Filtering

```python
# Create a new DataFrame containing only players who played more than 500 minutes
players_over_500_min_df = acc_basketball_df[acc_basketball_df['MP'] > 500]

# New DataFrame
print("Players who played more than 500 minutes:")
players_over_500_min_df
```

```python
# From this filtered DataFrame, determine the player with the highest total assists (AST)
top_assist_player = players_over_500_min_df.loc[players_over_500_min_df['AST'].idxmax()]
print(f"Player with the highest assists (played over 500 minutes): {top_assist_player['Player']}, Assists: {top_assist_player['AST']}")
```

```python
#  Who are the top 3 Assist Leads in the League?
top_3_assists = acc_basketball_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)

print("Top 3 Assist Leaders:")
print(top_3_assists)
```

```python
# Who are the top 3 Shot Blockers?
top_3_blockers = acc_basketball_df[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)

# Display the result
print("Top 3 Shot Blockers in the League:")
print(top_3_blockers)
```

### School Based Analysis

```python
# Group the players by School and calculate the total points scored by each school
total_points_by_school = acc_basketball_df.groupby('School')['PTS'].sum().reset_index()

# Results
print("Total points scored by each school:")
print(total_points_by_school)
```

```python
# Group by School and calculate total assists (AST) for each team
school_assists = acc_basketball_df.groupby('School')['AST'].sum().reset_index()

# Results
print("Total assists by each school:")
print(school_assists)
```

```python
# Sort the schools by total points scored and display the top 3 schools
school_points = acc_basketball_df.groupby('School')['PTS'].sum().reset_index()
school_points = school_points.sort_values(by='PTS', ascending=False).head(3)

# Results
print("Total points scored by each school:")
print(school_points)
```

### Extra Credit

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 5 players by points scored
top_5_points = acc_basketball_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Player', y='PTS', data=top_5_points)
plt.title('Top 5 Players by Points Scored')
plt.show()
```

```python
#  Correlation: Investigate whether there is any correlation between a player's field goal percentage (FG%) and their win shares (WS). Provide a scatter plot and calculate the correlation coefficient.

# Calculate the correlation coefficient
correlation = acc_basketball_df['FG%'].corr(acc_basketball_df['WS'])
print(f"Correlation coefficient between FG% and WS: {correlation}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FG%', y='WS', data=acc_basketball_df)

plt.title('Correlation Between FG% and Win Shares (WS)')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')

plt.show()
```
